# Import libraries

```python
# Import libraries
import yfinance as yf
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.gridspec as gridspec
import seaborn as sns
```

## Graphs

```python
# Define tickers and date range
tickers = ["SPY", "DIA", "QQQ", "IWM", "XLK"]
start_date = "2012-01-01"
end_date = "2024-01-01"

# Download and plot data for each ticker
for ticker in tickers:
    # Download data for the current ticker
    data = yf.download(ticker, start=start_date, end=end_date)
    data.reset_index(inplace=True)  # Reset index to have 'Date' as a column

    # Plot Adjusted Close Price
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Adj Close'], label=f'{ticker} Adjusted Close', color='blue')
    plt.title(f'Adjusted Close Price of {ticker} (2012-2023)')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.gcf().autofmt_xdate()  # Rotate x-axis labels for readability
    plt.show()

    # Plot Volume
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Volume'], label=f'{ticker} Volume', color='green')
    plt.title(f'Volume of {ticker} (2012-2023)')
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.legend()
    plt.gcf().autofmt_xdate()  # Rotate x-axis labels for readability
    plt.show()
```
